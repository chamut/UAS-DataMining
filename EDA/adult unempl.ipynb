```python
import pandas as pd
```

```python
df = pd.read_csv('D:/SEMESTER 6/Data Mining/Proyek UAS/Participants Datasets/SDG8 - Decent Work and Economic Growth/adult unemployment.csv')
```

```python
df
```

```python
#preprocess melt data
female = pd.melt(df, id_vars=['Countries'], value_vars=['2004F', '2005F', '2006F', '2007F', '2008F', '2009F', '2010F', '2011F', '2012F','2013F', '2014F'])
female.rename(columns = {'Countries':'country','variable': 'year', 'value': 'female'}, inplace = True)
female
```

```python
#preprocess melt data male
male = pd.melt(df, id_vars=['Countries'], value_vars=['2004M', '2005M', '2006M', '2007M', '2008M', '2009M', '2010M', '2011M', '2012M','2013M', '2014M'])
male.rename(columns = {'Countries':'country','variable': 'year', 'value': 'male'}, inplace = True)
male
```

```python
#gabung data melted male & female
female['male']= male['male']
female
```

```python
female['year'] = female['year'].replace({'F':''}, regex=True)
female
```

```python
adult_unem = female.copy()
adult_unem
```

```python
adult_unem.year.unique()
```

```python
adult_unem['country'] = adult_unem['country'].str.rstrip()
```

```python
adult_unem.info()
```

```python
adult_unem.year = adult_unem.year.str.strip()
adult_unem.year = adult_unem.year.astype('float').astype('Int32')
adult_unem.info()
```

```python
import matplotlib.pyplot as plt
def Countryinclude(country):
    #Slice and create the dataframes
    df_country = adult_unem.loc[adult_unem['country'] == country]
    #Starting to plot data from the country
    x = adult_unem['year']
    y = adult_unem['male']
    plt.plot(x,y, label=country, linestyle = '-', linewidth=2, alpha=1)
    plt.title('Adult Unemployment Rate')
    plt.xlabel('Year')
    plt.ylabel('Percentage')
    plt.legend(loc='best')
    
#Plot everything in one Figure
plot1 = plt.figure(1)

Countryinclude('Viet Nam')
```

```python
adult_unem.describe()
```

```python
negara = ['Brunei Darussalam', 'Cambodia', 'Indonesia', "Lao's PDR", 'Malaysia', 'Myanmar', 'Philippines', 'Singapore', 'Thailand', 'Viet Nam']
female= []
male = []

for n in negara:
    val = adult_unem[adult_unem['country']==n]['female'].mean()
    female.append(val)

for n in negara:
    val = adult_unem[adult_unem['country']==n]['male'].mean()
    male.append(val)

df_asean = pd.DataFrame({'Negara':['Brunei Darussalam', 'Cambodia', 'Indonesia', "Lao's PDR", 'Malaysia', 'Myanmar', 'Philippines', 'Singapore', 'Thailand', 'Viet Nam'], 'male': male, 'female': female})
df_asean
```

```python
df_asean.plot(x='Negara',
        kind='bar',
        stacked=False,
        figsize=(10,7),
        fontsize=12,
        title='Proporsi Unemployment Berdasarkan Gender')
plt.show()
```
