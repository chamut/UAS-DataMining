```python
#data modelling
train = (df_unemp_modelling[df_unemp_modelling.index < '2011-12-22'])
test = (df_unemp_modelling[df_unemp_modelling.index >= '2011-12-22'])
```

SARIMA

```python
# Define the p, d and q parameters to take any value between 0 and 2
p = d = q = range(0, 2)
```

```python
# Generate all different combinations of p, d and q triplets
pdq = list(itertools.product(p, d, q))
pdq
```

```python
# Generate all different combinations of seasonal p, q and q triplets
seasonal_pdq = [(x[0], x[1], x[2], 12) for x in list(itertools.product(p, d, q))]
```

```python
best_aic = np.inf
best_pdq = None
best_seasonal_pdq = None
temp_model = None
for param in pdq:   
    for param_seasonal in seasonal_pdq: 
        temp_model = SARIMAX(train,order=param,seasonal_order = param_seasonal,enforce_invertibility=False,
                             enforce_stationarity=False)
        results = temp_model.fit(disp=False)
        if results.aic < best_aic:
            best_aic = results.aic
            best_pdq = param
            best_seasonal_pdq = param_seasonal
print("Best SARIMA {} x {} model - AIC:{}".format(best_pdq,best_seasonal_pdq,best_aic)) 
```

```python
# using the best parameter in the model
model = SARIMAX(train,order=(0,1,1),seasonal_order=(1,1,1,12),enforce_invertibility=False,
                             enforce_stationarity=False)
model_fit = model.fit()
```

```python
predictions = model_fit.predict(start=len(train), end=len(train)+len(test)-1, dynamic=False)
predictions
```

```python
print('The first test value in Sales :',test.values[1],'and its Predicted value : ',predictions.values[1])
```
